Build a regression model.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Provide model output and an interpretation of the results. 

```python
model_df = pd.read_csv('../data/model_df.csv') 
```

```python
bikes_model_70 = model_df.copy()
bikes_dist_model_70 = bikes_model_70.drop(['percent_taken', 'yp_name_count','yp_price_avg','yp_rating_avg'], axis=1)
bikes_dist_model_70
```

```python
y = bikes_dist_model_70['fs_distance_avg']
X = bikes_dist_model_70.drop(['fs_distance_avg'], axis=1)
X = sm.add_constant(X)

model_70 = sm.OLS(y.astype(float), X.astype(float))
results_70 = model_70.fit()
print(results_70.summary())
```

```python
y = bikes_dist_model_70['fs_distance_avg']
X = bikes_dist_model_70.drop(['fs_distance_avg','bikes_total'], axis=1)
X = sm.add_constant(X)

model_70 = sm.OLS(y.astype(float), X.astype(float))
results_70 = model_70.fit()
print(results_70.summary())
```

```python
y = bikes_dist_model_70['fs_distance_avg']
X = bikes_dist_model_70.drop(['fs_distance_avg','bikes_total','fs_popularity_avg'], axis=1)
X = sm.add_constant(X)

model_70 = sm.OLS(y.astype(float), X.astype(float))
results_70 = model_70.fit()
print(results_70.summary())
```

```python
y = bikes_dist_model_70['fs_distance_avg']
X = bikes_dist_model_70.drop(['fs_distance_avg','bikes_total','fs_popularity_avg','fs_price_avg'], axis=1)
X = sm.add_constant(X)

model_70 = sm.OLS(y.astype(float), X.astype(float))
results_70 = model_70.fit()
print(results_70.summary())
```

```python
y = bikes_dist_model_70['fs_distance_avg']
X = bikes_dist_model_70.drop(['fs_distance_avg','bikes_total','fs_popularity_avg','fs_price_avg','fs_name_count'], axis=1)
X = sm.add_constant(X)

model_70 = sm.OLS(y.astype(float), X.astype(float))
results_70 = model_70.fit()
print(results_70.summary())
```

A coef of -27.3567 means that for each one-unit increase in fs_rating_avg, the fs_distance_avg (dependent variable) is expected to decrease by 27.3567 units, assuming all other variables in the model are held constant. A coef of 0.6296 means that for each one-unit increase in yp_distance_avg, the fs_distance_avg (dependent variable) is expected to increase by 0.6296 units, assuming all other variables in the model are held constant. Very loosely interpreting, a higher rated restaurant is likely closer to a bike station.

Although statistically significant with p-values under 0.05 and an Adj. R-squared of 0.592, this is not a reliable or helpful model. There would be high collinearity expected between the distance from station to a Foursquare restuarant, and distance from a station to a Yelp restaurant. If you already know the Yelp distance, it's not very helpful to predict the Foursquare distance.

```python
residuals_70 = results_70.resid
sns.histplot(residuals_70)
plt.show()
```

```python
results_70.params
```

```python
#distance_avg = b0 + b1*fs_rating + b2*yp_distance


b0 = results_70.params['const']
b1 = results_70.params['fs_rating_avg']
b2 = results_70.params['yp_distance_avg']

fs_rating = 6
yp_distance = 200
predicted_dist_70 = b0 + b1*fs_rating + b2*yp_distance
predicted_dist_70
```
